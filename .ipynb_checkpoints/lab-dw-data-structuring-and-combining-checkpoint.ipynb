```python
# Your code goes here
import pandas as pd
import numpy as np

url = "df_lab7.csv"
url1 = "https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file1.csv"
url2 ="https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file2.csv"
url3 = "https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file3.csv"

df1 = pd.read_csv(url1)
df2 = pd.read_csv(url2)
df3 = pd.read_csv(url3)
```

```python
display(df1.head())
display(df2.head())
display(df3.head())
```

```python
# Preparing the columns
df1.columns = list(map(lambda x: x.lower().replace(" ","_").strip(), df1.columns))
df1= df1.rename({'st' : 'state'}, axis = 1)
display(df1.columns)

df2.columns = list(map(lambda x: x.lower().replace(" ","_").strip(), df2.columns))
df2= df2.rename({'st' : 'state'}, axis = 1)
display(df2.columns)

df3.columns = list(map(lambda x: x.lower().replace(" ","_").strip(), df3.columns))
display(df3.columns)
```

```python
# Concatenating the 3 dataframes
df = pd.concat([df1, df2, df3])
df
```

```python
df.duplicated().value_counts()
```

```python
df = df.drop_duplicates()
df.duplicated().value_counts()
```

```python
df = df.dropna(axis = 0, how ="all")
df
```

```python
df['gender'].value_counts()
```

```python
def gender_clean(val):
    if val == "F":
        return "F"
    elif val == "M":
        return "M"
    elif val == "Femal":
        return "F"
    elif val == "female":
        return "F"    
    elif val == "Male":
        return "M"
    else:
        pass
    return val
```

```python
df.gender = df.gender.apply(gender_clean)
df.gender.value_counts()
```

```python
df.state.value_counts()
```

```python
def state_clean(val):
    if val == "Cali":
        val = "California"
    elif val == "WA":
        val = "Washington"
    elif val == "AZ":
        val = "Arizona"
    else:
        pass
    return val
```

```python
df.state = df.state.apply(state_clean)
df.state.value_counts()
```

```python
# In education, "Bachelors" could be replaced by "Bachelor"
df.education = df.education.str.replace("Bachelors","Bachelor")
df.education.value_counts()
```

```python
# In Customer Lifetime Value, delete the % character

df.customer_lifetime_value = df.customer_lifetime_value.str.replace("%", "")
```

```python
# In vehicle class, "Sports Car", "Luxury SUV" and "Luxury Car" could be replaced by "Luxury"
df.vehicle_class= df.vehicle_class.str.replace("Sports Car","Luxury").str.replace("Luxury SUV","Luxury").str.replace("Luxury Car","Luxury")

df.vehicle_class.value_counts()
```

```python
df["customer_lifetime_value"] = df["customer_lifetime_value"].apply(float)


```

```python
df.number_of_open_complaints.value_counts()
```

```python
type(df.number_of_open_complaints.iloc[-1])
```

```python
def complaints_clean(string):
    if type(string) == str:
        return int(string[2])
    else:
        pass

df.number_of_open_complaints = df.number_of_open_complaints.apply(complaints_clean)
```

```python
df.number_of_open_complaints.value_counts()
```

```python
df.notna().sum()
```

```python
df
```

```python
df = df.dropna(axis = 0, how ="all")
df
```

```python
df.gender= df.gender.fillna("N/A")
df
```

```python
df=df.reset_index(drop = True)
df
```

# Challenge 2: Structuring Data

In this challenge, we will continue to work with customer data from an insurance company, but we will use a dataset with more columns, called marketing_customer_analysis.csv, which can be found at the following link:

https://raw.githubusercontent.com/data-bootcamp-v4/data/main/marketing_customer_analysis_clean.csv

This dataset contains information such as customer demographics, policy details, vehicle information, and the customer's response to the last marketing campaign. Our goal is to explore and analyze this data by performing data cleaning, formatting, and structuring.

```python
# Your code goes here
url = "https://raw.githubusercontent.com/data-bootcamp-v4/data/main/marketing_customer_analysis_clean.csv"

data = pd.read_csv(url)

data
```

```python
data.notna().count()
```

```python
data = data.drop('unnamed:_0', axis=1)
data
```

```python
# looking at the data

for column in data.columns:
    display(data[column].value_counts())
```

```python
data.columns
```

```python
# 1/ You work at the marketing department and you want to know which sales channel brought the most sales in terms of total
# revenue. Using pivot, create a summary table showing the total revenue for each sales channel (branch, call center, web,
# and mail). Round the total revenue to 2 decimal points. Analyze the resulting table to draw insights.

pivot_data = data.pivot_table(index = 'sales_channel', columns = 'state', values = ['total_claim_amount'], aggfunc= sum)
pivot_data.round(2)
```

```python
pivot_data.round(2)/pivot_data.sum()
```

```python
# Jav: The proportion is about the same in all States. Agent generates close to 40% of the Total Claims, Branch 26-30%,
#      Call center 18-21% and finally web about 14-15%
```

```python
# 2/ Create a pivot table that shows the average customer lifetime value per gender and education level. Analyze the 
# resulting table to draw insights.

pivot_data2 = data.pivot_table(index = 'gender', columns = 'education', values = ['customer_lifetime_value'], aggfunc= 'mean')
pivot_data2.round(2)
```

```python
# On average Female have lower customer lifetime value, except for 'High School or Below' (the highest of all) and 'Master'
```
